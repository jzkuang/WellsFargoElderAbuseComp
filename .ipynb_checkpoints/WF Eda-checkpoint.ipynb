```python
to_datetime = ['PWD_UPDT_TS', 'PH_NUM_UPDT_TS', 'CUST_SINCE_DT','TRAN_TS',
               'TRAN_DT', 'ACTVY_DT']
for datetime in to_datetime:
  trainDf[datetime] = pd.to_datetime(trainDf[datetime], errors='coerce')
  #We're ignoring errors because some issues exist in time (June 31 in one of the columns!)
  #Coercing these values turns them into NaT (not a time)
trainDf.head()
```

```python
to_categorical = ['CARR_NAME', 'RGN_NAME', 'STATE_PRVNC_TXT', 'ALERT_TRGR_CD',
                  'DVC_TYPE_TXT', 'AUTHC_PRIM_TYPE_CD', 'AUTHC_SCNDRY_STAT_TXT',
                  'CUST_STATE','ACTN_CD','ACTN_INTNL_TXT','TRAN_TYPE_CD',
                  'FRAUD_NONFRAUD']

for category in to_categorical:
  trainDf[category] = trainDf[category].astype("category")
```

```python
trainDf.describe()
```

```python
#Visualize unique items in each categorical variable and 

for category in to_categorical:
  print("Category: " + category)
  print("nunique: " + str(trainDf[category].nunique()))
  print("----- Top -----")
  print(trainDf[category].value_counts())
  print("-------------------------------")
```

```python
trainDf.info()
```

```python
#Checking if ACTVY_DT is redudant, since nrow=14k, they are.
sum(trainDf['TRAN_DT'] == trainDf['ACTVY_DT'])
```

```python
redundant = ['ACTN_CD', 'TRAN_TYPE_CD','ACTN_INTNL_TXT','ACTVY_DT']
trainDf.drop(columns = redundant, inplace=True)
```

RGN_NAME

STATE_PRVNC_TXT

ALERT_TRGR_CD
  - Alert Trigger ???

DVC_TYPE_TXT
  - Device Type Text (Type of device used to send text)

AUTHC_PRIM_TYPE_CD

AUTHC_SCNDRY_STAT_TXT

CUST_ZIP
  - Customer ZIP

CUST_STATE
  - Customer State

PH_NUM_UPDT_TS
  - Phone number update timestamp (Time since last phone number change)

CUST_SINCE_DT
  - Customer since date

TRAN_TS

TRAN_DT
